```python
import pandas as pd
```

```python
# dictionary of lists
dict = {'name':["aparna", "pankaj", "sudhir", "Geeku"],
        'degree': ["MBA", "BCA", "M.Tech", "MBA"],
        'score':[90, 40, 80, 98]
}
```

```python
# creating a dataframe from a dictionary
df = pd.DataFrame(dict)
print(df)
```

```python
# Saving the DataFrame to a CSV file
df.to_csv('studdata.csv', index_label='roll')
```

```python
# Reading the DataFrame back from the CSV file
df_from_csv = pd.read_csv('studdata.csv', index_col='roll')
print(df_from_csv)
```
