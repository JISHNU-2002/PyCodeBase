# **Requests & JSON**

```python
import requests
import json
import random
import html
```

```python
url = "https://opentdb.com/api.php?amount=1"
end = ""
score_correct = 0
score_wrong = 0
```

```python
while end != "quit":
  r = requests.get(url)
  if(r.status_code != 200):
    end = input('quit or continue : ')
  else:
    valid = False
    num = 1
    data = json.loads(r.text)
    question = data['results'][0]['question']
    answers = data['results'][0]['incorrect_answers']
    ans = data['results'][0]['correct_answer']
    answers.append(ans)
    random.shuffle(answers)

    print(html.unescape(question))
    for answer in answers:
      print(html.unescape(str(num)+" - "+answer))
      num+=1

    while valid == False:
      user = input("option number : ")
      try:
        user = int(user)
        if(user <= 0 or user > len(answers)):
          print("invalid option")
        else:
          valid = True
      except:
        print("invalid option, use only numbers")

    if(user == str(answers.index(ans)+1)):
      print("correct")
      score_correct+=1
    else:
      print("wrong")
      score_wrong+=1

    print("correct : "+str(score_correct)+" wrong : "+str(score_wrong))
    end = input('quit or continue : ')
```
