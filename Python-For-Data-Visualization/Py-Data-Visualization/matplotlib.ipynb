# **matplotlib**


```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
import numpy as np
import time as t
```

```python
times = []
mistakes = 0
string = input("Enter the word : ")
n = int(input("No of try : "))
```

```python
while len(times)<n:
  start = t.time()
  word = input("Enter the word : ")
  end = t.time()
  times.append(end-start)

  if(string.lower() != word.lower()):
    mistakes += 1
    print("Wrong")
  else:
    print("Correct")
print("No of mistakes made : ",mistakes)
```

```python
x = [i for i in range(1,n+1)]
y = times
plt.plot(x, times)
plt.title('Time per Attempt')
plt.xlabel('attempt')
plt.ylabel('time in seconds')
plt.show
```

# Histograms

```python
# Generate random data
data = np.random.randn(1000)
# Create a histogram
plt.hist(data, bins=30, edgecolor='black')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Random Data')
plt.show()
```

density=True: Normalizes the histogram so that the area under the histogram integrates to 1, making it a density plot.

```python
# Create a customized histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Density Histogram of Random Data')

# Add a grid
plt.grid(True)

# Display the plot
plt.show()
```

# Pie Charts

```python
# Sample data
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [15, 30, 45, 10]  # Values representing each category
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']
explode = (0.1, 0, 0, 0)  # explode the 1st slice (Category A)

# Create the pie chart
# plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)

plt.title('Pie Chart Example')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Sample data
data = {
    'Categories': ['Category A', 'Category B', 'Category C', 'Category D'],
    'Values': [15, 30, 45, 10]
}
df = pd.DataFrame(data)

# Set the index to the categories for pie chart plotting
df.set_index('Categories', inplace=True)

# Create a pie chart using the kind keyword
# plt.figure(figsize=(8, 8))
df['Values'].plot(kind='pie',
                  autopct='%1.1f%%',
                  startangle=140,
                  # pctdistance=1.5,
                  shadow=True, colors=['gold', 'lightcoral', 'lightskyblue', 'lightgreen'])

plt.title('Pie Chart Example')
plt.ylabel('')  # Hides the y-label
plt.show()
```

Subplots in Box-plots

```python
# Sample data
data = {
    'Group A': [12, 15, 14, 10, 14, 13, 15, 16, 20, 19],
    'Group B': [22, 21, 20, 24, 25, 21, 20, 22, 30, 29],
    'Group C': [32, 31, 30, 29, 35, 33, 28, 30, 40, 38],
    'Group D': [42, 41, 45, 46, 43, 41, 45, 48, 46, 50]
}
df = pd.DataFrame(data)

# Create subplots
fig, axs = plt.subplots(2, 2)  # 2 row, 2 columns

# Box plot for Group A and Group B
df[['Group A', 'Group B']].plot(kind='box', ax=axs[0][0], grid=True, color={'boxes': 'blue', 'whiskers': 'orange', 'caps': 'green'})
axs[0][0].set_title('Box Plot for Group A and Group B')
axs[0][0].set_ylabel('Values')

# Box plot for Group B and Group C
df[['Group B', 'Group C']].plot(kind='box', ax=axs[0][1], grid=True, color={'boxes': 'red', 'whiskers': 'purple', 'caps': 'brown'})
axs[0][1].set_title('Box Plot for Group B and Group C')
axs[0][1].set_ylabel('Values')

# Box plot for Group A and Group c
df[['Group C', 'Group D']].plot(kind='box', ax=axs[1][0], grid=True, color={'boxes': 'blue', 'whiskers': 'orange', 'caps': 'green'})
axs[1][0].set_title('Box Plot for Group A and Group C')
axs[1][0].set_ylabel('Values')

# Box plot for Group A and Group c
df[['Group A', 'Group D']].plot(kind='box', ax=axs[1][1], grid=True, color={'boxes': 'blue', 'whiskers': 'orange', 'caps': 'green'})
axs[1][1].set_title('Box Plot for Group A and Group C')
axs[1][1].set_ylabel('Values')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Sample data
years = [2000, 2005, 2010, 2015, 2020]
values = [100, 200, 300, 400, 500]

# Create a DataFrame
df = pd.DataFrame({'Years': years, 'Values': values})

# Create a scatter plot using the kind keyword
ax = df.plot(kind='scatter', x='Years', y='Values', color='blue')

# Add annotations
for i, (year, value) in enumerate(zip(years, values)):
    ax.annotate(f'({year}, {value})', (year, value), textcoords="offset points", xytext=(0,10), ha='center')

# Add labels and title
plt.xlabel('Years')
plt.ylabel('Values')
plt.title('Growth Over Years')

# Show plot
plt.show()
```

# Waffle Charts

```python
import matplotlib.patches as patches
```

```python
# Define the data
data = {
    'Company A': 40,
    'Company B': 30,
    'Company C': 20,
    'Company D': 10
}

# Define the color for each category
colors = {
    'Company A': 'blue',
    'Company B': 'red',
    'Company C': 'green',
    'Company D': 'yellow'
}

# Create the waffle chart
fig, ax = plt.subplots()

# Number of rows and columns
n_rows = 10
n_cols = 10

# Plot each category
category_index = 0
current_pos = 0
for category, count in data.items():
    for i in range(count):
        row = current_pos // n_cols
        col = current_pos % n_cols
        rect = patches.Rectangle((col, n_rows - 1 - row), 1, 1, facecolor=colors[category], edgecolor='black', linewidth=1)
        ax.add_patch(rect)
        current_pos += 1
    category_index += 1

# Create legend
legend_handles = [patches.Patch(color=color, label=category) for category, color in colors.items()]
ax.legend(handles=legend_handles, loc='lower left')

# Set limits and aspect
ax.set_xlim(0, n_cols)
ax.set_ylim(0, n_rows)
ax.set_aspect('equal')

# Remove axes
ax.set_xticks([])
ax.set_yticks([])

plt.title('Market Share Waffle Chart')
plt.show()
```

```python
# Define the data
data = {
    'Asia Pacific': 32,
    'Australasia': 6,
    'Eastern Europe': 6,
    'Latin America': 8,
    'Middle East and Africa': 9,
    'North America': 10,
    'Western Europe': 29
}

# Define the color for each category
colors = {
    'Asia Pacific': '#17becf',
    'Australasia': '#ff7f0e',
    'Eastern Europe': '#2ca02c',
    'Latin America': '#d62728',
    'Middle East and Africa': '#9467bd',
    'North America': '#8c564b',
    'Western Europe': '#1f77b4'
}

# Create the waffle chart
fig, ax = plt.subplots(figsize=(10, 6))

# Number of rows and columns
n_rows = 10
n_cols = 10

# Plot each category
category_index = 0
current_pos = 0
for category, count in data.items():
    for i in range(count):
        row = current_pos // n_cols
        col = current_pos % n_cols
        rect = patches.Rectangle((col, n_rows - 1 - row), 1, 1, facecolor=colors[category], edgecolor='black', linewidth=1)
        ax.add_patch(rect)
        current_pos += 1
    category_index += 1

# Create legend
legend_handles = [patches.Patch(color=color, label=category) for category, color in colors.items()]
ax.legend(handles=legend_handles, loc='upper right', bbox_to_anchor=(1.3, 1))

# Set limits and aspect
ax.set_xlim(0, n_cols)
ax.set_ylim(0, n_rows)
ax.set_aspect('equal')

# Remove axes
ax.set_xticks([])
ax.set_yticks([])

plt.title('Where Are the Top 100 City Destinations?')
plt.show()
```
